```python
import re
import pickle

import pandas as pd

from sentence_transformers import SentenceTransformer
```

```python
def preprocess(string):
    string = string.lower()
    return re.sub(r"[\W_]+", " ", string)
```

```python
train_df = pd.read_json('data/train.jsonl', lines = True)
train_df["img"] = "data/" + train_df["img"]
train_df["text"] = train_df["text"].apply(preprocess)
```

```python
val_df = pd.read_json('data/dev.jsonl', lines = True)
val_df["img"] = "data/" + val_df["img"]
val_df["text"] = val_df["text"].apply(preprocess)
```

```python
transformer = SentenceTransformer('distilbert-base-nli-mean-tokens')
```

```python
embeds = []
for text in train_df["text"].values.tolist():
    embeds.append(transformer.encode([text]))
```

```python
val_embeds = []
for text in val_df["text"].values.tolist():
    val_embeds.append(transformer.encode([text]))
```

```python
pickle.dump(open('data/embeds.pickle', 'wb'), [i[0] for i in embeds])
```

```python
pickle.dump(open('data/val-embeds.pickle', 'wb'), [i[0] for i in val_embeds])
```
